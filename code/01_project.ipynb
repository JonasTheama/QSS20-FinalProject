```python
# What relationship types exist?
print("Relationship types:")
print(relationships['rel_type'].value_counts().head(10))

# What source investigations are represented?
print("\nSource IDs in officers:")
print(officers['sourceID'].value_counts())

# Country distribution for officers
print("\nTop officer countries:")
print(officers['countries'].value_counts().head(15))
```

```python
# drop unknowns, count
officer_countries = (
    officers[
        (~officers['countries'].isin(['Not identified', 'Not Identified'])) &
        (officers['countries'].notna())]
    ['countries']
    .str.split(';')          # some entries have more than one country
    .explode()
    .str.strip()
    .value_counts()
    .head(15)
)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(officer_countries.index[::-1], officer_countries.values[::-1], color='#004225')
ax.set_xlabel('Number of Officers', fontsize=12)
ax.set_title('Top 15 Countries of Origin for Officers\n(All Papers)', fontsize=14, fontweight='bold')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.savefig('viz1_officer_countries.png', dpi=150)
plt.show()
```

```python
#count jurisdictions
entity_jurisdictions = (
    entities[
        (entities['jurisdiction_description'].notna())
    ]
    ['jurisdiction_description']
    .value_counts()
    .head(15)
)

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(entity_jurisdictions.index[::-1], entity_jurisdictions.values[::-1], color='#004225')
ax.set_xlabel('Number of Registered Entities', fontsize=12)
ax.set_title('Top 15 Jurisdictions for Shell Company Registration\n(All Papers)', fontsize=14, fontweight='bold')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.savefig('viz2_entity_jurisdictions.png', dpi=150)
plt.show()
```
